#### 연습문제
- 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive)
- 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.
    - bins=[1,15,25,35,60,99]
    - labels=['미성년자','청년','중년','장년','노년']
    - 각 그룹별 비유의 전체 합은 1이 되어야 한다.

```python
# 생존
titanic_1=titanic[titanic.alive=='yes']
titanic_1
# 사망
titanic_0=titanic[titanic.alive=='no']
titanic_0
```

```python
bins=[1,15,25,35,60,99]
labels=['미성년자','청년','중년','장년','노년']
#생존자
data1=titanic_1.age
age1=pd.cut(data1,bins,labels=labels)
age1.value_counts(normalize=True)

#사망자
data2=titanic_0.age
age0=pd.cut(data2,bins,labels=labels)
age0.value_counts(normalize=True)
```
