### 함수로 구성
- get_movie(date)
    - date 인수를 사용해 url을 설정
    - date별로 영화정보(title,point)

```python
def get_movie(date):
    url_base='https://movie.naver.com'
    url_sub='/movie/sdb/rank/rmovie.naver?sel=cur&tg=0&date={date}'
    url=urljoin(url_base,url_sub)
    
    today=date
    result=urlopen(url.format(date=quote(today.strftime('%Y%m%d'))))
    
    soup=BeautifulSoup(result)
    
    # 랭킹 날짜 생성
    movie_date=today
    # 영화 제목 추출
    # 내포 for문을 이용한 영화 타이틀, 평점 추출
    movie_name=[s.a.text for s in soup.findAll('div',{'class','tit5'})]
    # 평점 추출
    movie_point=[s.text for s in soup.findAll('td',{'class','point'})]

    # 데이터프레임으로 반환
    return pd.DataFrame({'날짜':movie_date,'영화제목':movie_name,
                        '영화평점':movie_point})
```

```python
get_movie(date[17])
```

```python
# 설정기간 영화 정보 추출
movie_df=pd.DataFrame() # 빈 데이터프레임 생성

# tqdm 사용해서 진행상태 확인
for today in tqdm_notebook(date):
    movie_df=pd.concat([movie_df,get_movie(today)],axis=0,ignore_index=True)
```

### 시각화 작업
- 영화별평점의 합을 계산(평점의 합이 높은 영화를 추출)
- 추출된 영화에 대해서 평점의 추이를 확인할 수 있는 그래프 작성

```python
# 평점의 합을 계산하려면 영화평점은 수치형이어야 함
# 영화 평점 datatype 변환 : astype(float)
movie_df['영화평점']=movie_df['영화평점'].astype(float)
movie_df.info()
```

```python
# 영화별 point의 합 : pivot_table(data,index,columns,value,aggfunc=np.sum)
movie_unique=pd.pivot_table(movie_df,index=['영화제목'],aggfunc=np.sum)
movie_unique
```

```python
# 펑점의 합이 높은 순으로 정렬
movie_top=movie_unique.sort_values(by='영화평점',ascending=False)
movie_top
```

```python
# 씽2게더만 추출해서 평점 변화 확인
# 해당 영화만 필터링해서 확인
# df.query(조건)
tmp=movie_df.query('영화제목==["씽2게더"]')
tmp
```

```python
# 평점 변화 그래프
# 시계열 데이터이므로 line그래프
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['날짜'],tmp['영화평점'])
plt.grid()
```

```python
# 데이터 재배치 - 피봇테이블 사용(long-wide)
# 날짜별 영화의 point를 확인
# index=날짜,columns=영화제목, values=영화평점
movie_pivot=pd.pivot_table(movie_df,
                          index=['날짜'],
                          columns=['영화제목'],
                          values=['영화평점'])
```
