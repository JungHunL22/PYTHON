### 포털사이트 크롤링
- 소스내에서 특정 문자열(data)을 지칭하는 선택자 얻기
    - 크롬 개발자 도구 사용
- 전체 코드에서 수집하려고하는 데이터(태그)의 위치를 찾고 태그를 파싱한 후 필요데이터 추출

```python
from urllib.request import urlopen # 서버 요청/응답 패키지
import bs4 # 파싱패키지
```

```python
# 네이버 사이트의 기본 메뉴 문구 추출
url='http://www.naver.com'

# url로 요청 후 응답
html=urlopen(url)

# 파서객체 생성 - bs4 객체로 변환
bs_obj=bs4.BeautifulSoup(html,'html.parser')
```

```python
bs_obj
```

```python
test_menu=bs_obj.find('div',{'id':'gnb'})
test_menu
```

```python
#NM_FAVORITE > div.group_nav > ul.list_nav.type_fix
ul=bs_obj.find('ul',{'class':'list_nav type_fix'})
ul
```

```python
# 선택자로 찾기 클래스자체의 이름내에 띄어쓰기가 있을 경우 
# 자손태그로 볼 수 있기 때문에 공백이 있는 이름에는.으로 대체함
ul_test=bs_obj.select('#NM_FAVORITE > div.group_nav > ul.list_nav.type_fix')
ul_test # list형태로 반환
```

```python
type(ul) # bs4.element.Tag
# 문자열 아님
# 객체 안에 태그들이 여러개 존재
for u in ul:
    print(u)
```

```python
ul.findAll('li')
```

```python
lis=ul.findAll('li')
for li in lis:
    print(li.find('a')['href'])
    print(li.text)
```

#### 네이버 전체 메뉴 크롤링

```python
# 네이버 기본 메뉴는 ul > li.nav_item 형식으로 되어있음
lis=bs_obj.findAll('li',{'class':'nav_item'})
len(lis) # class 속성 값이 nav_item인 li태그 추출(16개 존재)
```

```python
for li in lis:
    a_tag=li.find('a')
    print(a_tag.text,":",a_tag['href'])
```

### 수집한 데이터를 csv로 저장
1. 항목별로 list에 저장
2. 항목들을 dict로 구성
3. dick를 데이터프레임으로 생성
4. 데이터프레임을 csv로 저장

```python
# 네이버 메뉴 정보 저장
# 빈 리스트 생성
menu=[]
url=[]
```

```python
for li in lis:
    a_tag=li.find('a')
    menu.append(a_tag.text)
    url.append(a_tag['href'])
```

```python
menu, url
```

```python
import pandas as pd
df=pd.DataFrame({'메뉴':menu, 'URL':url})
df
```

```python
df.to_csv('crawl_data/naver.menu.csv')
```

### 네이버 뉴스섹션메뉴와 섹션별 url추출

```python
from urllib.request import urlopen
import bs4
import pandas as pd
```

```python
url='http://news.naver.com'
html=urlopen(url)
bs_obj=bs4.BeautifulSoup(html,'html.parser')
```

```python
# 네이버 뉴스 섹션메뉴 태그 확인(개발자 도구)
# body > section > header > div.Nlnb._float_lnb > div > div > div.Nlnb_left._lnb_scroll > div > div > ul
# selector가 너무 길어 적정하지 않아 보임
# 직접 확인한 태그와 클래스 속성 사용
# ul 태그의 class : Nlnb_menu_list
uls=bs_obj.findAll('ul',{'class':'Nlnb_menu_list'})
len(uls)
```

```python
ul=bs_obj.find('ul',{'class':'Nlnb_menu_list'})
ul
```

```python
lis=ul.findAll('li')
len(lis)
```

```python
for li in lis:
    a_tag=li.find('a')
    print(a_tag.text,':',a_tag['href'])
```

```python
# li태그를 이용해서 추출
lis=bs_obj.findAll('li',{'class':'Nlist_item'})
len(lis)
```

```python
for li in lis:
    a_tag=li.find('a')
    print(a_tag.text)
    print(a_tag['href'])
```

```python
# 수집데이터 df로 구성 후 저장
section=[]
link=[]
```

```python
for li in lis:
    a_tag=li.find('a')
    section.append(a_tag.text)
    link.append(a_tag['href'])
```

```python
col_dict={'section':section,'link':link}
news_section_df=pd.DataFrame(col_dict)
```

```python
news_section_df
```

```python
news_section_df.to_csv('crawl_data/naver_news_section.csv')
```
