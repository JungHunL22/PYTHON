6. 출생년도를 4개의 카테고리로 분류하여 연령대 필드를 생성하시오. 
- (청년(19에서 29세); 중년(30에서 49세); 장년(50에서 64세); 노년(65세 이상))
- 2020년 나이를 기준으로 경계(bins)는 직접 설정 하시오

````python
bins=[1938,1956,1972,1992,2002]
labels=['노년','장년','중년','청년']
연령대=pd.cut(raw_data.생년,bins=bins,labels=labels)
raw_data['연령대']=연령대
````

9. raw_data_open df의 컬럼명을 아래와 같이 변경하시오.
- strd_yymm : 수집년월

````python
raw_data_open.rename(columns={'strd_yymm' : '수집년월',
                              'card_open_strd_yymm' : '1차수집년월',
                              'deto_id' : '사용자ID',
                              'inst_id' : '금융기관ID',
                              'card_open_yymm' : '개설년월'},inplace=True)
````

10. 카드개설년월필드에 대해서 카드개설연도 카드개설월로 필드를 분리하시오

````python
raw_data_open['카드개설연도']=raw_data_open['개설년월'].astype(str).str.slice(start=0,stop=4).astype(int)
raw_data_open['카드개설월']=raw_data_open['개설년월'].astype(str).str.slice(start=4,stop=6).astype(int)
````

````python
raw_data_open.head()
````

11. card_open_rson_nm 필드는 81이면 신용카드 83이면 신용체크카드를 의미한다 
 개설유형 필드를 새로 생성하고 해당 코드에 맞는 유형으로 데이터를 변환하시오


````python
raw_data_open.loc[raw_data_open['card_open_rson_nm']==81, '개설유형'] = '신용카드'
raw_data_open.loc[raw_data_open['card_open_rson_nm']==83, '개설유형'] = '신용체크카드'
````

````python
raw_data_open.head()
````

12. card_typ_nm 필드는 개인카드(1)인지 개인기업카드(2)인지를 나타낸다
- 카드유형 필드를 새로 추가하고 card_typ_nm 값에 따라 개인/기업으로 나타내시오

````python
raw_data_open.loc[raw_data_open['card_typ_nm']==1, '카드유형'] = '개인'
raw_data_open.loc[raw_data_open['card_typ_nm']==2, '카드유형'] = '기업'
````

````python
raw_data_open
````

13. card_typ_nm 필드와 card_open_rson_nm 필드는 제거하시오

````python
raw_data_open.drop(['card_typ_nm','card_open_rson_nm'],axis=1,inplace=True)
````

````python
raw_data_open.head()
````

14. 카드개설연도 필드에 대해서 각 연대별로 카테고리를 구분하여 개설연대 필드를 생성하시오.
- 1990년대, 2000년대, 2010년대

````python
bins=[1989,2000,2010,2020]
labels=['1990년대','2000년대','2010년대']
raw_data_open['개설연대']=pd.cut(raw_data_open['카드개설연도'],bins,labels=labels)
raw_data_open.head()
````

15. 카드 개설 df와 차주정보 df를 memberID를 기준으로 병합하여 새로운 df로 저장하고 확인하시오
- inner_join 

````python
fin_data=pd.merge(raw_data,raw_data_open,on='사용자ID',how='inner')
````

````python
fin_data.head()
len(fin_data)
````

16. 카드 개설 인원을 청년, 중년, 장년, 노년으로 비교할 수 있도록 아래와 같은 그래프로 표현하시오.

17. 시간의 흐름에 따른 카드 개설 인원의 추이를 확인할 수 있는 아래와 같은 그래프를 표현하시오.


````python
df2=pd.DataFrame(fin_data.groupby('개설년월').size(),
                columns=['카드개설인원']).reset_index()
plt.figure(figsize=(10,6))
plt.plot(df2.개설년월,df2.카드개설인원)
````

18. fin_data df를 이용하여 각 연대별 발급건수를 표현하는 피봇테이블을 생성하고 해당 테이블을 사용하여 각 연대별 신용카드 발급 추이를 나타내는 아래와 같은 그래프를 그리시오


````python
df3=pd.pivot_table(fin_data,
                    index=['개설연대'],
                    values=['사용자ID'],
                    aggfunc='count')
df3
````

````python
plt.figure(figsize=(10,10))
plt.plot(df3,color='green',ls='--',marker='s',ms=10)
````

![그래프](문제18.png)

19. 신용카드의 각 연도별 발급건수를 계산하고 연도별 오름차순으로 정렬하시오.
- 결과를 변수에 저장하시오


````python
df4=pd.DataFrame(pd.pivot_table(fin_data,index=['카드개설연도'],aggfunc='size')).reset_index()
df4.columns=['카드개설연도','발급건수']
df4=df4.sort_values(by='카드개설연도')
df4
````

20. 19번에서 저장한 변수의 데이터를 활용하여 연도별 
- 개설건수에 대하여 아래와 같은 시계열 그래프로 표현하시오.


````python
x=[i for i in range(1995,2020,5)]
x
plt.figure(figsize=(10,6))
plt.plot(df4.카드개설연도,df4.발급건수)
plt.xticks(x)
````

![그래프](문제20.png)

21. 성별,연령대로 그룹화하여 카드 발급 건수를 표현하는 피봇테이블을 작성하고 변수에 저장하시오


````python
df5=pd.DataFrame(pd.pivot_table(fin_data,index=['성별'],columns=['연령대'],aggfunc='size'))

df5=df5[['노년','장년','중년','청년']].sort_values(by='노년')
````

````python
x=[0,1]
df5.plot.bar(grid=True,figsize=(10,6))
plt.legend(['(발급건수, 노년)','(발급건수, 장년)','(발급건수, 중년)','(발급건수, 청년)'], title = "None,None",loc="upper right")
plt.xticks(x,['남','여'])
````
